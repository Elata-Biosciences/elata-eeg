```python
print('Load data and get properties\n')

fname = '2025-03-21_16-10_gain24_boardAds1299_vref4.5.csv'
df = pd.read_csv(f"../recordings/{fname}")

board, gain, Fs, vref, avss, resolution, vmid = util.get_board_attributes(fname)
channel_count = len([i for i in list(df.columns) if i.find('_raw_sample') !=-1])
df = util.add_time_and_raw_voltage_columns(df, fname)

print(f"\nvref={vref}V, avss={avss}V, gain={gain}, sample_rate={Fs}hz, vmid={vmid}, resolution={resolution} bit")
print('\nchannel count=', channel_count)
print('\ndf columns=', list(df.columns), '\n')
```

```python
from scipy.signal import butter, iirnotch, filtfilt


def onboard_filter(signal, bandpass_order, lowcut, highcut, notch_q):
    # bandpass
    nyq = 0.5 * Fs
    b, a = butter(bandpass_order, [lowcut / nyq, highcut / nyq], btype='band')
    signal = filtfilt(b, a, signal)
    
    nyq = 0.5 * Fs
    b, a = iirnotch(60.0 / nyq, notch_q)
    signal = filtfilt(b, a, signal)
    
    nyq = 0.5 * Fs
    b, a = iirnotch(50.0 / nyq, notch_q)
    signal = filtfilt(b, a, signal)
    
    return signal
```

```python
print('GUI relies on the onboard filter. Tweak the onboard driver filter to remove 60hz noise')
```

```python
# === Parameters ===
trim_time = 4.0  # seconds to trim from start
start_index = int(trim_time * Fs)

# Use these to tweak the driver filter
sim_order = 4
sim_lowcut = 1
sim_highcut = 40
sim_notch_q = 30

# Signals
displayed_gui_voltage = df['ch2_voltage'].values[start_index:]
ch2_raw_voltage = df['ch2_raw_voltage'].values[start_index:]
sim_gui_voltage = onboard_filter(ch2_raw_voltage, sim_order, sim_lowcut, sim_highcut, sim_notch_q)
N = len(sim_gui_voltage)


# Tranforms
# gui_voltage = gui_voltage * np.hamming(N)
displayed_gui_fft_vals = np.fft.fft(displayed_gui_voltage)
displayed_gui_fft_freqs = np.fft.fftfreq(N, d=1/Fs)
displayed_gui_fft_vals = displayed_gui_fft_vals[displayed_gui_fft_freqs > 0]
displayed_gui_fft_freqs = displayed_gui_fft_freqs[displayed_gui_fft_freqs > 0]
displayed_gui_power = np.abs(displayed_gui_fft_vals) ** 2 / N  # Power spectrum (V²)

# sim_gui_voltage = sim_gui_voltage * np.hamming(N)
sim_gui_fft_vals = np.fft.fft(sim_gui_voltage)
sim_gui_fft_freqs = np.fft.fftfreq(N, d=1/Fs)
sim_gui_fft_vals = sim_gui_fft_vals[sim_gui_fft_freqs > 0]
sim_gui_fft_freqs = sim_gui_fft_freqs[sim_gui_fft_freqs > 0]
sim_gui_power = np.abs(sim_gui_fft_vals) ** 2 / N  # Power spectrum (V²)
```

```python
%matplotlib ipympl

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# EEG bands
eeg_bands = {
    "Delta (0.5–4 Hz)": (0.5, 4),
    "Theta (4–8 Hz)": (4, 8),
    "Alpha (8–13 Hz)": (8, 13),
    "Beta (13–30 Hz)": (13, 30),
    "Gamma (30–45 Hz)": (30, 45),
}
colors = ['gray', 'purple', 'green', 'orange', 'red']

# === Plot ===

# Top: Displayed GUI FFT
ax1.plot(displayed_gui_fft_freqs, np.abs(displayed_gui_fft_vals) / N, label="Displayed FFT", color='b')
for (label, (f_min, f_max)), color in zip(eeg_bands.items(), colors):
    ax1.axvspan(f_min, f_max, color=color, alpha=0.2, label=label)
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(dict(zip(labels, handles)).values(), dict(zip(labels, handles)).keys())
ax1.set_ylabel("Magnitude")
ax1.set_title("Displayed GUI Signal FFT")
ax1.grid()

# Bottom: Simulated Signal FFT
ax2.plot(sim_gui_fft_freqs, np.abs(sim_gui_fft_vals) / N, label="Simulated FFT", color='g')
for (label, (f_min, f_max)), color in zip(eeg_bands.items(), colors):
    ax2.axvspan(f_min, f_max, color=color, alpha=0.2, label=label)
handles, labels = ax2.get_legend_handles_labels()
ax2.legend(dict(zip(labels, handles)).values(), dict(zip(labels, handles)).keys())
ax2.set_xlabel("Frequency (Hz)")
ax2.set_ylabel("Magnitude")
ax2.set_title("Simulated Signal FFT")
ax2.grid()

ax1.set_xlim(0, 80)  # <-- Add this line
ax2.set_xlim(0, 80)  # <-- Add this line
fig.tight_layout()
```
